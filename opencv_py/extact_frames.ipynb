```python
import cv2
import os
```

```python
# file lists
file_list = os.listdir('data/240p')
video_list = []
for f in file_list:
    video_list.append(f.split('_')[0])
    
rates = ['720p', '240p']
print(video_list)
print(rates)
```

```python
# extracting configuration
interval_second = 30
```

```python
# target
for video_name in video_list:
    for rate in rates[:1]:
        # data
        src_folder = f'data/{rate}/'
        file_name = video_name + f'_{rate}.mp4'
        train_folder = f'train/{rate}/'
        path = src_folder + file_name
        # read video
        video = cv2.VideoCapture(path) # 引数がファイルパス
        print('file path:', path)
        width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH)) # 動画の画面横幅
        height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT)) # 動画の画面縦幅
        frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT)) # 総フレーム数
        frame_rate = int(video.get(cv2.CAP_PROP_FPS)) # 
        print('width:',  width)
        print('height:',  height)
        print('frame_count:',  frame_count)
        print('frame_rate:',  frame_rate)
        
        interval = frame_rate * interval_second

        for f in range(frame_count):
            ret, frame = video.read()
            if f % interval == 0:
                if rate == '720p':
                    frame = frame[:, 1:-1, :] ##(720, 1280, 3) =>(720, 1278, 3)
                savename = train_folder + video_name +f'_{f}.jpg' 
                print(savename)
                cv2.imwrite(savename, frame) 
```
