```python
import pandas as pd
import numpy as np
```

```python
# definition
c = np.array([4, 8, 10, 0, 0, 0])
A = np.array([[1, 1, 1, 1, 0, 0],
              [3, 4, 6, 0, 1, 0],
              [4, 5, 3, 0, 0, 1]])
b = np.array([20, 100, 100])

N_list = np.array([0, 1, 2])
B_list = np.array([3, 4, 5])
```

```python
cnt = 0
while True:
    print(f"-------{cnt}------")
    cnt += 1
    print("N_list: ", N_list)
    print("B_list: ", B_list)
    
    N = A[:, N_list]
    B = A[:, B_list]
    cn = c[N_list]
    cb = c[B_list]

    # rearanging dictionary
    N_bar = np.dot(np.linalg.inv(B), N)
    b_bar = np.dot(np.linalg.inv(B), b)
    cn_bar = cn - np.dot(N.T, np.dot(np.linalg.inv(B).T, cb))
    
    print("cn_bar: ", cn_bar)
    z = np.dot(cb, b_bar)
    print(f"z:{z:.2f}")
    if (cn_bar <= 0).sum() == len(cn_bar):
        break
    
    k = np.argmax(cn_bar)
    ak = N_bar[:, k]
    i = np.argmin(b_bar / ak)

    print(f"k:{k}")
    print(f"i:{i}")

    N_list[k], B_list[i] = B_list[i], N_list[k]

```
