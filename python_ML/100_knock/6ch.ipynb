```python
INPUT_DIR = "サンプルコード_20201021/6章/"
```

```python
import os
import pandas as pd
```

```python
os.listdir(INPUT_DIR)
```

```python
df_fact = pd.read_csv(INPUT_DIR + "tbl_factory.csv", index_col=0)
df_tran = pd.read_csv(INPUT_DIR + "tbl_transaction.csv", index_col=0)
df_ware = pd.read_csv(INPUT_DIR + "tbl_warehouse.csv", index_col=0)
df_cost = pd.read_csv(INPUT_DIR + "rel_cost.csv", index_col=0)
```

```python
join_data = pd.merge(df_tran, df_cost, left_on=["ToFC", "FromWH"], right_on=["FCID", "WHID"], how="left")
join_data.head()
```

```python
join_data = pd.merge(join_data, df_fact, left_on="ToFC", right_on="FCID", how="left")
join_data.head()
```

```python
join_data = pd.merge(join_data, df_ware, left_on="FromWH", right_on="WHID", how="left")
join_data = join_data[["TransactionDate", "Quantity", "Cost", "ToFC", "FCName", "FCDemand", "FromWH", "WHName", "WHSupply", "WHRegion"]]
join_data.head()
```

```python
kanto = join_data.loc[join_data["WHRegion"]=="関東"]
kanto.head()
```
