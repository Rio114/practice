```python
import os
import pandas as pd
import numpy as np

INPUT_DIR = "サンプルコード_20201021/8章/"
os.listdir(INPUT_DIR)
```

```python
df_link = pd.read_csv(INPUT_DIR + 'links.csv').set_index("Unnamed: 0")
df_link.head()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

# num = len(df_link.index)
for col in df_link.columns:
    node_no = col.strip("Node")
    G.add_node(str(node_no))

for row in df_link.index:
    for col in df_link.columns:
        if df_link[row][col] == 1:
            G.add_edge(str(row.strip("Node")), str(col.strip("Node")))

nx.draw_networkx(G, node_color="k", edge_color="k", font_color="w")
plt.show()
```

```python
import numpy as np
```

```python
def determine_link(percent):
    rand_val = np.random.rand()
    if rand_val<=percent:
        return 1
    else:
        return 0
    
def simulate_percolation(df_link, actives, percent_percolation):
    for i, row in enumerate(df_link.index):
        if actives[i] == 1:
            for j, col in enumerate(df_link.columns):
                if df_link[row][col] == 1:
                    if determine_link(percent_percolation) == 1:
                        actives[j] = 1
    return actives
```

```python
percent_percolation = 0.05
T_NUM = 100
num = len(df_link.index)
actives = np.zeros(num)
actives[0] = 1
timeSeries = []
for t in range(T_NUM):
    actives = simulate_percolation(df_link, actives, percent_percolation)
    timeSeries.append(actives.copy())
```

```python
def active_node_coloring(actives):
    colors = []
    for i in range(len(timeSeries[t])):
        if timeSeries[t][i] == 1:
            colors.append("r")
        else:
            colors.append("k")
            
    return colors
```

```python
for t in range(100):
    if (t + 1) % 10 == 0:
        nx.draw_networkx(G, font_color="w", node_color=active_node_coloring(timeSeries[t]))
        plt.show()
```

```python
num_timeSeries = []
for s in timeSeries:
    num_timeSeries.append(s.sum())

plt.plot(num_timeSeries)
plt.show()
```
