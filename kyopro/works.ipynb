```python
import numpy as np
import time
```

```python
fib_list = [1, 1]
```

```python
def calc_fib(num):
    fib_list = [1, 1]
    for i in range(num)[2:]:
        fib_list.append(fib_list[i-1]+fib_list[i-2])
    return fib_list[-1]
```

```python
for i in range(100):
    start = time.time()
    out = calc_fib(i)
    duration = time.time() - start
    print(i, out, duration)
```
